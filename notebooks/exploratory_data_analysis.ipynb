# Exploratory Data Analysis (EDA) of Bilibili Video Data

This notebook performs an exploratory data analysis on the scraped Bilibili video data stored in our MySQL database. The goal is to understand the data's underlying structure, distributions, and relationships, which will inform our subsequent feature engineering and modeling steps.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
import os

# Set plot style
sns.set_style('whitegrid')
```

## 1. Load Data from MySQL Database

```python
# --- Database Configuration ---
DB_USER = os.getenv("DB_USER", "root")
DB_PASSWORD = os.getenv("DB_PASSWORD") # Assumes you've set this environment variable
DB_HOST = os.getenv("DB_HOST", "localhost")
DB_PORT = os.getenv("DB_PORT", "3306")
DB_NAME = os.getenv("DB_NAME", "bilibili_data")

# Construct the database URL
DATABASE_URL = f"mysql+pymysql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"

if not DB_PASSWORD:
    print("FATAL: DB_PASSWORD environment variable is not set.")
else:
    try:
        engine = create_engine(DATABASE_URL)
        
        # SQL query to select all data from the 'videos' table
        query = "SELECT * FROM videos;"
        
        # Load data into a pandas DataFrame
        df = pd.read_sql(query, engine)
        
        print("Successfully loaded data from the database.")
        
    except Exception as e:
        print(f"An error occurred while connecting to the database or loading data: {e}")
```
