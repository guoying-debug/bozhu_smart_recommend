```python
import pandas as pd

# Define the path to the data file, relative to the project root.
file_path = '../src/bilibili_scraper/output.json'

try:
    # Scrapy's -o command appends to existing files, which can corrupt the JSON structure if run multiple times.
    # It creates a 'JSON lines' format, where each line is a separate JSON object.
    # We use pd.read_json with lines=True to correctly parse this format.
    df = pd.read_json(file_path, lines=True, encoding='utf-8')
    
    # Display the first few rows to verify the data is loaded correctly.
    print("First 5 rows of the dataset:")
    print(df.head())
    
    # Display a concise summary of the DataFrame to check data types and non-null counts.
    print("\nDataFrame Info:")
    df.info()

except FileNotFoundError:
    print(f"Error: The file was not found at {file_path}")
except Exception as e:
    # Catching other potential errors, like parsing issues if the file is malformed for other reasons.
    print(f"An unexpected error occurred: {e}")
```

```python
# --- Data Cleaning and Preprocessing ---

# 1. Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# 2. Convert Unix timestamp to datetime objects
# The 'publish_time' is in seconds, so we specify unit='s'
df['publish_time'] = pd.to_datetime(df['publish_time'], unit='s')

# 3. Ensure numeric columns are of the correct type (e.g., int)
numeric_cols = ['view_count', 'like_count', 'coin_count', 'favorite_count', 'share_count', 'comment_count']
for col in numeric_cols:
    # astype(int) is used for clean conversion
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)

# 4. Handle missing or empty values in 'description' and 'tags'
# Fill NaN in 'description' with an empty string
df['description'] = df['description'].fillna('')
# 'tags' and 'top_comments' are lists, so we can check for empty lists, but they seem to be handled correctly by scrapy.
# We'll assume for now they are valid empty lists if no tags/comments were scraped.

# Display the info again to see the changes
print("\nDataFrame Info after cleaning:")
df.info()

# Display the first 5 rows with the cleaned data
print("\nFirst 5 rows after cleaning:")
print(df.head())
```
